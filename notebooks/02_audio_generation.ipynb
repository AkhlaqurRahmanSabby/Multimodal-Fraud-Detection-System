**Import Audio Libraries**

```python
import pandas as pd
import os
import time
from gtts import gTTS
from IPython.display import Audio
```

**Load Cleaned Data**

```python
processed_path = "../data/processed/cleaned_transcripts.csv"
df = pd.read_csv(processed_path)
df.head()
```

**Test Audio Generation**

```python
audio_dir = "../data/processed/audio"
os.makedirs(audio_dir, exist_ok=True)

sample_text = df['text'][0]
print(f"Generating audio for: {sample_text[:50]}...")

# Generate the audio using Google TTS
tts = gTTS(text=sample_text, lang='en')
test_audio_path = f"{audio_dir}/test_sample.wav"
tts.save(test_audio_path)
Audio(test_audio_path)
```

**Generate Full Audio**

```python
audio_dir = "../data/processed/audio"
os.makedirs(audio_dir, exist_ok=True)

audio_paths = []
print("Starting full audio generation (this may take a few minutes)...")

for index, row in df.iterrows():
    file_path = f"{audio_dir}/sample_{index}.wav"
    
    if not os.path.exists(file_path):
        tts = gTTS(text=row['text'], lang='en')
        tts.save(file_path)
        time.sleep(1.5)  # Pauses for 1.5 seconds to prevent rate limiting
        
    audio_paths.append(file_path)
    
    if (index + 1) % 100 == 0:
        print(f"Processed {index + 1} / {len(df)} files...")

df['audio_path'] = audio_paths
print("Full dataset generated successfully!")
```

**Save Multimodal Dataset**

```python
multimodal_path = "../data/processed/multimodal_dataset.csv"
df.to_csv(multimodal_path, index=False)
print(f"Successfully saved multimodal dataset to: {multimodal_path}")

df.head()
```
