**Import and Locate Files**

```python
import pandas as pd
import os
import re
```

```python
scam_file_path = "../data/raw/English_Scam.txt"
non_scam_file_path = "../data/raw/English_NonScam.txt"

print("Scam file exists:", os.path.exists(scam_file_path))
print("Non-scam file exists:", os.path.exists(non_scam_file_path))
```

**Read Text into Memory**

```python
with open(scam_file_path, 'r', encoding='utf-8') as f:
    scam_lines = f.readlines()

with open(non_scam_file_path, 'r', encoding='utf-8') as f:
    non_scam_lines = f.readlines()

print(f"Total scam transcripts: {len(scam_lines)}")
print(f"Total non-scam transcripts: {len(non_scam_lines)}")
```

```python
print("--- First 3 Scam Lines ---")
for i in range(3):
    print(scam_lines[i].strip())

print("\n--- First 3 Non-Scam Lines ---")
for i in range(3):
    print(non_scam_lines[i].strip())
```

**Build Structured Dataframe**

```python
scam_data = [{'text': line.strip(), 'label': 1} for line in scam_lines if line.strip()]
non_scam_data = [{'text': line.strip(), 'label': 0} for line in non_scam_lines if line.strip()]

df = pd.DataFrame(scam_data + non_scam_data)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df.head()
```

**Clean Transcript Text**

```python
df['text'] = df['text'].apply(lambda x: re.sub(r'^\d+\.\s*', '', x))
df.head()
```

**Save Cleaned Data**

```python
processed_path = "../data/processed/cleaned_transcripts.csv"
df.to_csv(processed_path, index=False)
print(f"Successfully saved {len(df)} transcripts to {processed_path}")
```
